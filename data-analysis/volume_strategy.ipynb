```python
from lib.scripts import get_dataframe
```

```python
df = get_dataframe("TQBR", "TGKB", "2023-01-01", "2023-08-30", "100")
```

```python
df
```

```python
median = df[['VOLUME']].median()
```

```python
df[['VOLUME/VOLUME_MEDIAN']] = (df[['VOLUME']] / median)
df[['CLOSE/OPEN']] = (df[['CLOSE']].values / df[['OPEN']].values)
df[['HIGH/LOW']] = (df[['HIGH']].values / df[['LOW']].values)
prepared_df = df[['TRADEDATE', 'OPEN', 'CLOSE', 'VOLUME', 'VOLUME/VOLUME_MEDIAN', 'CLOSE/OPEN', 'HIGH/LOW']]
```

```python
prepared_df
```

```python
buy_df = prepared_df[(prepared_df['VOLUME/VOLUME_MEDIAN'] > 1) & (prepared_df['HIGH/LOW'] > 1)]
```
